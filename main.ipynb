# Gait-cycle Based Approach Demo

```python
from Utils.scanner import read_beijing_dataset
```

## Data Pre-Processing

```python
# Load Beijing Xuanhe dataset (limit to 1 patient for quick testing)
database = read_beijing_dataset(patient_no_limit=1)
database.downsample_to_50hz()
database.remove_high_freq_noise()
database.generate_pre_fog_label(pre_fog_duration=3)
database.normalize_features()
```

## Construct Gait-cycle Based Segments

```python
# Construct and plot gait-cycle based segments
gaitSegments, gaitSegmentsLabel = database.get_segs_and_labels(segmentLength=64)
database.patients[0].trails[0].plot_gaitSegments()
```

## Construct Sliding Windows

```python
# Construct and plot sliding windows
slidingWindows, slidingWindowsLabels = database.get_slidingWindows_and_labels(windowSize=64, windowStride=48)
database.patients[0].trails[0].plot_slidingWindows()
```
